```python
# needed imports
import numpy as np
import matplotlib.pyplot as plt
from bmd import *
from bmd_biv import *
import scipy.io as sio
from numba import jit
import matplotlib
from matplotlib.gridspec import GridSpec
import matplotlib.patches as patches
```

```python
# sets default font parameters
matplotlib.rcParams['mathtext.fontset'] = 'stix'
matplotlib.rcParams['font.family'] = 'STIXGeneral'
matplotlib.rcParams.update({'font.size': 17})
```

```python
# color list
colors1 = ['#ff0000', '#ff5555', '#ff8888', '#ffbbbb', '#ffdddd']
```

```python
# sets figure/axis/grid sizes
fig, ax = plt.subplots(figsize=(23, 5.5), dpi=300)
ax.axis('off')
gs = GridSpec(nrows=23, ncols=80)

# box size parameters for subfigure a)
xw = 0.58
xd = 0.66
yh = 0.26
yxn = 0.02
yyn = 0.0046
xxn = 0.0046
xyn = 0.02

# colored boxes for subfigure a)
rect = patches.Rectangle((xw/2, 4/5*yh), xw/2, yh/5, facecolor = colors1[0] ,fill=True, lw=0)
ax.add_patch(rect)
for i in range(2):
    rect = patches.Rectangle((xw*(2*i+1)/4, 3/5*yh), xw/4, yh/5, facecolor = colors1[1] ,fill=True, lw=0)
    ax.add_patch(rect)
for i in range(4):
    rect = patches.Rectangle((xw*(2*i+1)/8, 2/5*yh), xw/8, yh/5, facecolor = colors1[2] ,fill=True, lw=0)
    ax.add_patch(rect)
for i in range(8):
    rect = patches.Rectangle((xw*(2*i+1)/16, 1/5*yh), xw/16, yh/5, facecolor = colors1[3] ,fill=True, lw=0)
    ax.add_patch(rect)
for i in range(16):
    rect = patches.Rectangle((xw*(2*i+1)/32, 0/5*yh), xw/32, yh/5, facecolor = colors1[4] ,fill=True, lw=0)
    ax.add_patch(rect)

# lines for subfigure a)
ax.plot([0,xw],yh*(1/5)*np.ones(2),':k')
ax.plot([0,xw],yh*(2/5)*np.ones(2),':k')
ax.plot([0,xw],yh*(3/5)*np.ones(2),':k')
ax.plot([0,xw],yh*(4/5)*np.ones(2),':k')

# S-set labels for subfigure a)
snx = 0.034
sny = 0.0055
ax.text(-snx, (4.5/5)*yh-sny, '$\mathcal{S}_1$', fontsize=22)
ax.text(-snx, (3.5/5)*yh-sny, '$\mathcal{S}_2$', fontsize=22)
ax.text(-snx, (2.5/5)*yh-sny, '$\mathcal{S}_3$', fontsize=22)
ax.text(-snx, (1.5/5)*yh-sny, '$\mathcal{S}_4$', fontsize=22)
ax.text(-snx, (0.5/5)*yh-sny, '$\mathcal{S}_5$', fontsize=22)

# axis labels for subfigure a)
ax.text(-xxn, -xyn, 'a', fontsize=22)
ax.text(xw/2-xxn, -xyn, 'x', fontsize=22)
ax.text(xw-xxn, -xyn, 'b', fontsize=22)

# box for subfigure a)
rect = patches.Rectangle((0, 0), xw, yh, edgecolor = 'black', facecolor = 'none',fill=False, lw=1)
ax.add_patch(rect)

# box for subfigure b)
rect = patches.Rectangle((xd, 0), xw, yh, edgecolor = 'black', facecolor = 'none',fill=False, lw=1)
ax.add_patch(rect)

# lines for subfigure b)
ax.plot([xd,xd+xw*(4.5/5)],[0,yh],'--k')
ax.plot([xd,xd+xw],yh*0.1*np.ones(2),':k')
ax.plot([xd,xd+xw],yh*0.9*np.ones(2),':k')
ax.plot((xd+0.1*(4.5/5)*xw)*np.ones(2),[0,yh],':k')
ax.plot((xd+0.9*(4.5/5)*xw)*np.ones(2),[0,yh],':k')

# colored boxes for subfigure b)
rect = patches.Rectangle(((xd+0.5*(4.5/5)*xw), 0.1*yh), 0.8*(4.5/5)*xw/2, (0.9-0.1)*yh/2, facecolor = colors1[0] ,fill=True, lw=0)
ax.add_patch(rect)
for i in range(2):
    rect = patches.Rectangle(((xd+(0.1+0.8*(1/4+1/2*i))*(4.5/5)*xw), (0.1+i*0.8/2)*yh), 0.8*(4.5/5)*xw/4, (0.9-0.1)*yh/4, facecolor = colors1[1] ,fill=True, lw=0)
    ax.add_patch(rect)    
for i in range(4):
    rect = patches.Rectangle(((xd+(0.1+0.8*(1/8+1/4*i))*(4.5/5)*xw), (0.1+i*0.8/4)*yh), 0.8*(4.5/5)*xw/8, (0.9-0.1)*yh/8, facecolor = colors1[2] ,fill=True, lw=0)
    ax.add_patch(rect)    
for i in range(8):
    rect = patches.Rectangle(((xd+(0.1+0.8*(1/16+1/8*i))*(4.5/5)*xw), (0.1+i*0.8/8)*yh), 0.8*(4.5/5)*xw/16, (0.9-0.1)*yh/16, facecolor = colors1[3] ,fill=True, lw=0)
    ax.add_patch(rect)  
for i in range(16):
    rect = patches.Rectangle(((xd+(0.1+0.8*(1/32+1/16*i))*(4.5/5)*xw), (0.1+i*0.8/16)*yh), 0.8*(4.5/5)*xw/32, (0.9-0.1)*yh/32, facecolor = colors1[4] ,fill=True, lw=0)
    ax.add_patch(rect)

# axis labels for subfigure b)
ax.text(xd-yxn, yh*0.9-yyn, 'b', fontsize=22)
ax.text(xd-yxn-0.008, yh*0.5, 'y', fontsize=22, rotation=90, va='center')
ax.text(xd-yxn, yh*0.1-yyn, 'a', fontsize=22)
ax.text(xd-yxn, 0-yyn, '0', fontsize=22)
ax.text(xd-xxn, 0-xyn, '0', fontsize=22)
ax.text((xd+0.1*(4.5/5)*xw)-xxn, -xyn, 'a', fontsize=22)
ax.text(xw/2+xd-xxn, 0-xyn, 'x', fontsize=22)
ax.text((xd+0.9*(4.5/5)*xw)-xxn, -xyn, 'b', fontsize=22)
ax.text(xw/2+xd-xxn+0.02, 0.21-xyn, '$y=x$', fontsize=22)

# position parameters
lxn = 0.03
lyn = 0.02
# Labels for letters a) - b)
ax.text(0-lxn, yh+lyn,'a)', fontsize=27)  
ax.text(xd-lxn, yh+lyn,'b)', fontsize=27)
ax.text(0+0.06, yh+0.01,'Visualization of Family of Sets $\mathcal{S}$ from Definition 1', fontsize=20) # title for a)
ax.text(0+xd+0.06, yh+0.01,'$y=x$ Constructed from Sum of $\mathcal{S}_i$ Indicator Functions ', fontsize=20) # title for b)

plt.show() # show plot
```
